```python
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

# Coordenadas del contorno exterior del polígono
exterior_coords = [(0, 0), (0, 10), (10, 10), (10, 0)]

# Coordenadas de los agujeros dentro del polígono
hole_coords = [[(2, 2), (2, 3), (3, 3), (3, 2)], [(7, 7), (7, 8), (8, 8), (8, 7)]]

# Crear el polígono con agujeros
polygon = Polygon(exterior_coords, hole_coords)

# Crear una figura y ejes de Matplotlib
fig, ax = plt.subplots()

# Graficar el contorno exterior del polígono
exterior_x, exterior_y = polygon.exterior.xy
ax.plot(exterior_x, exterior_y, color='blue')

# Graficar los agujeros dentro del polígono
for hole in polygon.interiors:
    hole_x, hole_y = hole.xy
    ax.plot(hole_x, hole_y, color='red')

# Configuraciones adicionales del gráfico
ax.set_aspect('equal', 'box')
ax.set_title('Polígono con agujeros')
plt.grid(True)
plt.show()


```

```python
from shapely.geometry import Point
import matplotlib.pyplot as plt

# Crear un punto en las coordenadas (2, 3)
punto = Point(2, 3)

# Extraer las coordenadas x e y del punto
x, y = punto.x, punto.y

# Crear un gráfico
plt.figure()
plt.plot(x, y, 'ro')  # 'ro' significa puntos rojos
plt.title('Gráfico de Punto')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.grid(True)
plt.show()
```

```python
from shapely.geometry import LineString
import matplotlib.pyplot as plt

# Crear una línea con puntos de (1, 2) a (4, 5)
linea = LineString([(1, 2), (4, 5)])

# Extraer las coordenadas x e y de los puntos de la línea
x, y = linea.xy

# Crear un gráfico
plt.figure()
plt.plot(x, y, 'b-')  # 'b-' significa una línea azul sólida
plt.title('Gráfico de Línea')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.grid(True)
plt.show()
```

```python
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

# Coordenadas del contorno exterior del polígono principal
exterior_coords = [(0, 0), (0, 10), (10, 10), (10, 0)]

# Coordenadas del polígono más pequeño (modificado para ser más grande y cerca del borde del polígono principal)
hole_coords = [(0.3, 0.3), (0.3, 9.7), (9.7, 9.7), (9.7, 0.3)]

# Crear el polígono principal y el polígono más pequeño
polygon_principal = Polygon(exterior_coords)
polygon_pequeño = Polygon(hole_coords)

# Extraer las coordenadas x e y del polígono principal
x_principal, y_principal = polygon_principal.exterior.xy

# Extraer las coordenadas x e y del polígono más pequeño
x_pequeño, y_pequeño = polygon_pequeño.exterior.xy

# Crear un gráfico con ejes cuadrados
plt.figure(figsize=(6, 6))  # Ajustar el tamaño de la figura para que sea cuadrada
plt.plot(x_principal, y_principal, 'g-')  # Polígono principal en azul
plt.fill(x_principal, y_principal, 'green',alpha=0.5)  # Rellenar el polígono principal

plt.plot(x_pequeño, y_pequeño, 'w-')  # Polígono más pequeño en rojo
plt.fill(x_pequeño, y_pequeño, 'white')  # Rellenar el polígono más pequeño

plt.title('Gráfico de Polígonos')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.gca().set_aspect('equal', adjustable='box')  # Hacer que los ejes tengan la misma escala
plt.grid(True)
plt.show()

```
