## Paso 2: Definir Reglas para Detectar Oro Fraudulento
Vamos a definir las reglas específicas para cada rubro basadas en las señales de fraude mencionadas anteriormente. 
### Reglas para Minería:

```python
def detectar_fraude_mineria(row):
    mina_id = row['Mina_ID']
    proporcion_oro = row['Proporcion_Oro']
    proporcion_esperada = inicial_minas.loc[mina_id, 'Muestras - Porcentaje Oro']
    
    if abs(proporcion_oro - proporcion_esperada) > 5:
        return 'Sí'
    else:
        return 'No'

# Aplicar la función para detectar fraude en la tabla de eventos de minería
eventos_mina['fraudulento'] = eventos_mina.apply(detectar_fraude_mineria, axis=1)
```

### Reglas para Joyería:

```python
def detectar_fraude_joyeria(row):
    segunda_procedencia = row['Segunda_Procedencia_Nombre']
    
    if pd.isnull(segunda_procedencia) or segunda_procedencia == 'Desconocido' or eventos_joyero[eventos_joyero['Segunda_Procedencia_Nombre'] == segunda_procedencia].shape[0] > 1:
        return 'Sí'
    else:
        return 'No'

# Aplicar la función para detectar fraude en la tabla de eventos de joyería
eventos_joyero['fraudulento'] = eventos_joyero.apply(detectar_fraude_joyeria, axis=1)
```

### Reglas para Cooperativas:

```python
def detectar_fraude_cooperativas(row):
    cooperativa_id = row['Cooperativa_ID']
    oro_entregado = row['Oro_GR']
    
    if cooperativa_id not in inicial_cooperativas['ID'].tolist() or inicial_barequeros[inicial_barequeros['Cooperativa_ID'] == cooperativa_id]['Cedula'].duplicated().any() or oro_entregado > 35:
        return 'Sí'  # Devuelve 'Sí' si hay problemas con los barequeros o entregas excesivas
    else:
        return 'No'  # Devuelve 'No' si no hay problemas detectados
```

## Paso 3: Generar Tablas Resumen por Rubro
Ahora, vamos a generar tablas resumen que clasifiquen el oro por cada rubro y cuánto de este es fraudulento.

```python
# Tabla resumen por Minería
resumen_mineria = eventos_mina.groupby('Mina_ID')['fraudulento'].value_counts(normalize=True).unstack().fillna(0)

# Tabla resumen por Joyería
resumen_joyeria = eventos_joyero.groupby('Joyero_ID')['fraudulento'].value_counts(normalize=True).unstack().fillna(0)

# Tabla resumen por Cooperativas
resumen_cooperativas = eventos_barequeros.groupby('Cooperativa_ID')['fraudulento'].value_counts(normalize=True).unstack().fillna(0)

# Mostrar las tablas resumen
print("Resumen de Oro Fraudulento por Minería:")
print(resumen_mineria)

print("\nResumen de Oro Fraudulento por Joyería:")
print(resumen_joyeria)

print("\nResumen de Oro Fraudulento por Cooperativas:")
print(resumen_cooperativas)
```

## Paso 4: Visualización de Resultados por Rubro
Finalmente, vamos a utilizar Matplotlib y Seaborn para crear gráficos que muestren el porcentaje de oro fraudulento por proveedor en cada rubro.

### Gráfico para Minería:

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=resumen_mineria.index, y=resumen_mineria['Sí'], hue=resumen_mineria.index, dodge=False)
plt.title('Porcentaje de Oro Fraudulento por Mina')
plt.xlabel('Mina')
plt.ylabel('Porcentaje de Oro Fraudulento')
plt.legend(title='Fraudulento', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Gráfico para Minería:

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=resumen_joyeria.index, y=resumen_joyeria['Sí'], hue=resumen_joyeria.index, dodge=False)
plt.title('Porcentaje de Oro Fraudulento por Joyero')
plt.xlabel('Joyero')
plt.ylabel('Porcentaje de Oro Fraudulento')
plt.legend(title='Fraudulento', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Gráfico para Cooperativas:

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=resumen_cooperativas.index, y=resumen_cooperativas['Sí'], hue=resumen_cooperativas.index, dodge=False)
plt.title('Porcentaje de Oro Fraudulento por Cooperativa')
plt.xlabel('Cooperativa')
plt.ylabel('Porcentaje de Oro Fraudulento')
plt.legend(title='Fraudulento', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
